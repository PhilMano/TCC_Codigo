Quantidade total das colunas de Rx(dB)

```python
quantidadeTotalRx = datasetfiltrado['Rx(dB)'].count()
```

Tratamento de conversão dos valores de string para float, com intuito de comparação posterior dos valores maiores e iguais a 0 da coluna Rx(dB)

```python
datasetfiltrado['Rx(dB)'] = datasetfiltrado['Rx(dB)'].astype(float)
```

Quantidade das colunas em que Rx(dB) são maiores ou iguais a zero

```python
quantidadeMaioresquezero = datasetfiltrado[datasetfiltrado['Rx(dB)'] > 0]['Rx(dB)'].count()
quantidadeMenoresIguaisazero = datasetfiltrado[datasetfiltrado['Rx(dB)'] <= 0]['Rx(dB)'].count()
```

```python
#Cálculo da porcentagem de Rx(dB)
```

```python
porcentagem = (quantidadeMaioresquezero /quantidadeTotalRx)*100
```
